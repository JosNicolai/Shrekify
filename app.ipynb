```python
import streamlit as st
from fastai.vision.all import *
```

```python
img = PILImage.create("download.jpg")
img.thumbnail((192,192))
img
```

```python
learn = load_learner("export.pkl")
```

```python
learn.predict(img)
```

```python
# select categories and select path
categories = ('shrek dragon','shrek donkey','shrek', "lord farquaad", "puss in boots")
def classify_image(img):
    pred, idx, probs = learn.predict(img)
    return dict(zip(categories, map(float, probs))) # streamlit might not be able to recognise tensors
```

```python
classify_image(img)
```
